**Exploratory Data Analysis**

```python
df['Year'].value_counts().plot(kind='bar')
```

```python
df['State'].value_counts().plot(kind='bar')
```

```python
df['Vehicle_Category'].value_counts().plot(kind = 'pie', autopct ='%.2f%%')
```

```python
df['Vehicle_Type'].value_counts().plot(kind = 'pie', autopct ='%.2f%%')
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df, x='Year', y='EV_Sales_Quantity', marker='o')
plt.title('EV Sales Quantity Over Years')
plt.xlabel('Year')
plt.ylabel('EV_Sales_Quantity')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='State', y='EV_Sales_Quantity',
            estimator=sum, ci=None)
plt.title('EV Sales Quantity by State')
plt.xlabel('State')
plt.ylabel('EV_Sales_Quantity')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='State', hue='Vehicle_Category')
plt.title('Vehicle Category Distribution by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Vehicle Category')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Month_Name', y='EV_Sales_Quantity', estimator=sum, ci=None)
plt.title('EV Sales Quantity by Month')
plt.xlabel('Month')
plt.ylabel('EV Sales Quantity')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
